```
import pysbd
seg = pysbd.Segmenter(language="en", clean=False)
SegmentedList = seg.segment(data)

Num = len(SegmentedList)
print('Number_of_Sentences: ',Num)

# for i in range(len(SegmentedList)):
#   res = len(SegmentedList[i].split())

```

```
import re
import string

#Code to find the number of Uppercase words
 
for i in range(len(SegmentedList)):
  Count = 0
  for j in SegmentedList[i]:
    if (j.isupper()):
        Count = Count + 1
  print('Number of uppercase words in sentence:'+str(i), Count)
#print('Total Number of uppercase words :' + str(Count))

#Code to find the Number of Numerics

for i in range(len(SegmentedList)):
  Count = 0
  for j in SegmentedList[i]:
    if j.isdigit() == True:
        Count = Count + 1
  
  print('Number of numerics in sentence:'+str(i), Count)

def remove(list):
  pattern = '[0-9]'
  list = [re.sub(pattern,'',i) for i in list]
  return list

#code for removing the numbers from list of strings --data cleaning 
list = SegmentedList
NumberFreeList = remove(list)


# code to find the number of special characters
 
for i in range(len(NumberFreeList)):
  Count = 0
  for j in NumberFreeList[i]:
    if j != ' ':
      if j.isdigit() == False:
        if j.isalpha() == False:
          Count = Count + 1

  print('Number of Special characters in Sentence:'+str(i), Count)


#code to remove punctuations from list of strings

x = [''.join(c for c in s if c not in string.punctuation) for s in NumberFreeList]


#code to clear the empty strings from list
y = [string for string in x if string != ""]
# print(y)

# print(NewList)

for i in range(len(y)):
  WordsCount = 0
  res = len(y[i].split())
  WordsCount = WordsCount + res
 
  print ("The number of words in sentence:"+str(i), WordsCount) 
```

```

for i in range(len(y)):
  CharCount = 0
  count = len(y[i]) - y[i].count(' ')
  CharCount = CharCount + count

  print("The number of characters in sentence:"+str(i), CharCount)

# code to calculate average word length
averages = []
for i in range(len(y)):
  words = y[i].split()
  if len(words)==0:
    continue
  average = sum(len(word) for word in words)/len(words)
  averages.append(average)

WordLenAverage = sum(averages)/len(averages)
print("The Average word length: "  + str(WordLenAverage))
```

```
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))


count = 0
for i in range(len(y)):
  sentence = y[i]
  word_tokens = word_tokenize(sentence)
  for w in word_tokens:
    if w in stop_words:
      count = count + 1
    
    
print("Number of stopwords: " + str(count))

```

```
#1.2

#Code to convert the text into Lower Case
LowerCaseList = []
for i in y:
  LowerCaseList.append(i.lower())

#Code to remove the stopwords from the List
FilteredSentence = []
for i in range(len(LowerCaseList)):
  sentence = LowerCaseList[i]
  sentence = ' '.join([word for word in sentence.split() if word not in stopwords.words("english")])
  FilteredSentence.append(sentence)

# FilteredSentence
```

```
#code for removing the frequent words
import collections 
wordfreq = []
wordList = []
words_ = []
FinalWords = []
lowfreq = []
FinalWordsList = []
for i in FilteredSentence:
  wordList.append(word_tokenize(i))
for j in wordList:
  for k in j:
    words_.append(k)
counter = collections.Counter(words_)

#code to remove the top 10 high frequency words
high_freq = []
for a in counter.most_common(10):
  high_freq.append(a[0])
for y in FilteredSentence:
  b = []
  split = word_tokenize(y)
  for i in split:
    if i not in high_freq:
      b.append(i)
  FinalWords.append(' '.join(b))

#code to remove the top 10 rare words
for x in counter.most_common()[:-11:-1]:
  lowfreq.append(x[0])
for z in FinalWords:
  a = []
  splits = word_tokenize(z)
  for c in splits:
    if c not in lowfreq:
      a.append(c)
  FinalWordsList.append(' '.join(a))
  # FinalWordsList=list(filter(None,FinalWordsList))
  
FinalWordsList = [string for string in FinalWordsList if string != ""] # for removing empty string values from a list
# FinalWordsList = [string for string in FinalWordsList if string != "-"]
# FinalWordsList
```

```
import nltk
from nltk.tokenize import word_tokenize
StemWords=[]
tokenizedList = []                                    #code for tokenization
for i in range(len(FinalWordsList)):
# Tokenizing strings in list of strings 
  tokenizedList.append(word_tokenize(FinalWordsList[i]))

tokenizedList = [string for string in tokenizedList if string != "-"]
tokenizedList = [string for string in tokenizedList if string != ","]
# tokenizedList = [string for string in tokenizedList if string != " "]


from nltk.stem import PorterStemmer                   #code for stemming
porter = PorterStemmer()

StemWords=[]
for i in tokenizedList:
  sent=''
  # print(i)
  for j in i:
    sent=sent+" "+porter.stem(j)
    sent.strip()
  StemWords.append(sent)
# StemWords
StemWords = [string for string in StemWords if string != ""]
# StemWords
```

```
nltk.download('wordnet')    #code for Lemmatization
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizedWords = []
lemmatizer = WordNetLemmatizer()

for i in StemWords:
  sent=''
  for j in i:
    sent=sent+""+lemmatizer.lemmatize(j)
    sent.strip()
  lemmatizedWords.append(sent)
# lemmatizedWords
```

```
#1.3       Code for moving to a csv file
import pandas as pd
df = pd.DataFrame(lemmatizedWords)
csv_data = df.to_csv('out.csv')

#1.4       #code for calculating term frequency of all the terms
wordList1 = []
words_1 = []
for i in lemmatizedWords:
  wordList1.append(word_tokenize(i))
for j in wordList1:
  for k in j:
    words_1.append(k)
counter = collections.Counter(words_1)

counter
```

## 2.2 Write a Python Program to extract all the years from the following sentence.

sentence = "The 2010s were a dramatic decade, filled with ups and downs, more than 1000 stroies have happened. As the decade comes to a close, Insider took a look back at some of the biggest headline-grabbing stories, from 2010 to 2019. The result was 119 news stories that ranged from the heartwarming rescue of a Thai boys' soccer team from a flooded cave to the divisive election of President Donald Trump."

```
# Write your code here

import re
def extract_year(sentence):
  return re.findall(r'(2\d{3})', sentence)

sentence = "The 2010s were a dramatic decade, filled with ups and downs, more than 1000 stroies have happened. As the decade comes to a close, Insider took a look back at some of the biggest headline-grabbing stories, from 2010 to 2019. The result was 119 news stories that ranged from the heartwarming rescue of a Thai boys' soccer team from a flooded cave to the divisive election of President Donald Trump."

print(extract_year(sentence))
```
